```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
## Define the vocabulary size
voc_size=10000
```

```python
one_hot_repr=[one_hot(words,voc_size) for words in sent]
one_hot_repr
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import pad_sequences

sent_len = 8
embedded_docs = pad_sequences(one_hot_repr,padding="post",maxlen=sent_len)
embedded_docs
```

```python
dim=10 #featrue representation
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_len))
model.compile('adam','mse')
```

```python
#ende to end deeplearning projects using rnn
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense
```

```python
max_features = 10000 #vocabulary size
(x_train,y_train),(x_test,y_test) = imdb.load_data(num_words=max_features)

# Print the shape of the data
print(f'Training data shape: {x_train.shape}, Training labels shape: {y_train.shape}')
print(f'Testing data shape: {x_test.shape}, Testing labels shape: {y_test.shape}')
```

```python
sample_review = x_train[0]
sample_label = y_train[0]

print(f"sample review (as int):{sample_review}")
print(f"sample label: {sample_label}")
```

```python
#mapping of word index

word_index = imdb.get_word_index()
reverse_wrd_index = {value:key for key,value in word_index.items()}
reverse_wrd_index
```

```python
from tensorflow.keras.preprocessing import sequence

max_len = 500
x_train = sequence.pad_sequences(x_train, maxlen=max_len)
x_test = sequence.pad_sequences(x_test, maxlen=max_len)
```

```python
x_train[0]
```

```python
#train rnn model
model = Sequential()
model.add(Embedding(max_features,128,input_length=max_len)) #embedding layers
model.add(SimpleRNN(128,activation="relu"))
model.add(Dense(1,activation='sigmoid'))
model.build(input_shape=(None, max_len))
# model.summary()
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
#create an instance of eraly stopping
from tensorflow.keras.callbacks import EarlyStopping
earlystopping = EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
earlystopping
```

```python
#train the model with earlystopping
history=model.fit(x_train,y_train,epochs=10,batch_size=32,validation_split=0.2,callbacks=[earlystopping])
history
```

```python
model.save("simple_rnn_imdb.h5")
```

```python
# Step 1: Import Libraries and Load the Model
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import load_model
```

```python

# Load the IMDB dataset word index
word_index = imdb.get_word_index()
reverse_word_index = {value: key for key, value in word_index.items()}
```

```python
# Load the pre-trained model with ReLU activation
model = load_model('simple_rnn_imdb.h5')
model.summary()
```

```python
model.get_weights()
```

```python
# Step 2: Helper Functions
# Function to decode reviews
def decode_review(encoded_review):
    return ' '.join([reverse_word_index.get(i - 3, '?') for i in encoded_review])

# Function to preprocess user input
def preprocess_text(text):
    words = text.lower().split()
    encoded_review = [word_index.get(word, 2) + 3 for word in words]
    padded_review = sequence.pad_sequences([encoded_review], maxlen=500)
    return padded_review
```

```python
### Prediction  function

def predict_sentiment(review):
    preprocessed_input=preprocess_text(review)

    prediction=model.predict(preprocessed_input)

    sentiment = 'Positive' if prediction[0][0] > 0.5 else 'Negative'

    return sentiment, prediction[0][0]

```

```python
# Step 4: User Input and Prediction
# Example review for prediction
example_review = "This movie was fantastic! The acting was great and the plot was thrilling."

sentiment,score=predict_sentiment(example_review)

print(f'Review: {example_review}')
print(f'Sentiment: {sentiment}')
print(f'Prediction Score: {score}')
```
